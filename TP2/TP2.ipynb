## **4- C’est quoi pandas series, labels, avec exemple ?**

## **10- Comment analyser les données avec Pandas ?**
- visualiser les données: s, il existe plusieurs méthodes de visualisation des données, nous verrons certains d’entre eux :
    > la méthode **head()** retourne les en-têtes et un nombre spécifié de lignes, à partir du **haut**.\
    > La méthode **tail()** retourne les en-têtes et un nombre spécifié de lignes, à partir du **bas**.\
    > La méthode **info()**, qui vous donne plus d’informations sur l’ensemble de données.

    **voir un exemple ci-dessous**

````python
import pandas as pd

Df = pd.read_csv("data.csv")

print(Df.head(7)) # Affichera les 7 premières lignes de la DataFrame
print("\n\n",Df.head()) # si le nombre de lignes n’est pas spécifié, la méthode head() retournera les 5 premières lignes

print("\n", Df.tail(2)) #affiche les 2 dernières lignes du DataFrame
print("\n", Df.tail()) #affiche les 5 dernières lignes du DataFrame

print("\n les info de Df sont :\n")
print(Df.info())
````

## **11- Nettoyage des données avec pandas**
Le nettoyage des données signifie corriger les données erronées dans votre DataSet.

Les données pourraient être :

- Cellules vides
- Données dans le mauvais format
- Données fausses
- Doublons

Nous verrons des exemples de chacun d’entre eux :

#### `Nous verrons un exemple de DataSet, qui contient toutes les données de nettoyage`avec le fichier **Nettoyage.csv**

````python
import pandas as pd

NDF = pd.read_csv("Nettoyage.csv")

print(NDF)
````

> - __Cellules vides__ (__Date__ à la ligne 22, et __Calories__ aux lignes 18 et 28).
>
> - **Données dans le mauvais format** (__Date__ à la ligne 26).
>
> - __Données fausses__ (__Durée__ à la ligne 7).
> - __Doublons__ (lignes 11 et 12).

### Donc pour mieux nettoyage : 
- **Cellules vies** : utilisez la méthode **`df.dropna()`**,cette méthode supprimera les lignes, Ou **`fillna()`** Ou pour remplacer les valeurs vides, utilisez cette méthode.
- **Données dans la mauvais format** : Pour convertir dans un format correct, utilisez cette méthode **`to_datetime()`**
- **Données fausses :** Utiliser ce 
```python 
df.loc[7, 'Duration'] = 45 
```
pour remplacer les valeurs par des valeurs correctes 
- __Doublons__ : Pour supprimer les doublons, utilisez la méthode **`drop_duplicates().`**
