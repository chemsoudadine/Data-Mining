## **4-programme python comment éliminer les caractères non alphabétiques à partir d'un ensemble de text et définir ?** 
- soit une séquence de lettres (attention aux accents!). 
- soit un chiffre 
- soit un symbole de ponctuation 
- une date ex : 12/12/2012 
---

**Solution :**
Je vais proposer ce texte : 
> `Aujourd’hui j’ai exécuté plus de 20 fois, pour confirmer ce code, qu’il travaille ! ou non ? Ce jour de : 29/10/2021`

- [x] Séquence de lettres : ex (exécuté , **é**)
- [x] Chifre __20__
- [x] ymbole de ponctuation (**? , : ’ !**)
- [x] Date : **29/10/2021**


````plaintext
import nltk
from nltk.corpus import stopwords
import re

TextFr =" Aujourd’hui j’ai exécuté plus de 20 fois, pour confirmer ce code, qu’il travaille ! ou non ? Ce jour de : 29/10/2021 "
print(TextFr)

TextFr = TextFr.lower()
TextFr = re.sub(r'\w' ,' ', TextFr)
TextFr = re.sub(r's+',' ', TextFr)

print(TextFr)

MotsVide = set(stopwords.words("french"))
print(MotsVide)

MotsToken = {}
Token = nltk.word_tokenize(TextFr)

for Tkn in Token :
  if Tkn not in MotsVide :
    if Tkn not in MotsToken.keys():
      MotsToken[Tkn]=1
    else:
        MotsToken[Tkn]+=1

print(MotsToken)
print(Token)

print(TextFr)
````

````plaintext
import nltk
import numpy as np
import random
import string
import bs4 as bs
import urllib.request
import re

OurText = " Beans. I was trying to explain to somebody as we were flying in, that’s corn. That’s beans. And they were very impressed at my agricultural knowledge. Please give it up for Amaury once again for that outstanding introduction. I have a bunch of good friends here today, including somebody who I served with, who is one of the finest senators in the country, and we’re lucky to have him, your Senator, Dick Durbin is here. I also noticed, by the way, former Governor Edgar here, who I haven’t seen in a long time, and somehow he has not aged and I have. And it’s great to see you, Governor. I want to thank President Killeen and everybody at the U of I System for making it possible for me to be here today. And I am deeply honored at the Paul Douglas Award that is being given to me. He is somebody who set the path for so much outstanding public service here in Illinois. Now, I want to start by addressing the elephant in the room. I know people are still wondering why I didn’t speak at the commencement. "

print(OurText) # Affiche notre Text

TknText = nltk.sent_tokenize(OurText)
for i in range(len(TknText)):
    TknText[i] = TknText[i].lower()
    TknText[i] = re.sub(r'\W', ' ', TknText[i])
    TknText[i] = re.sub(r'\s+', ' ', TknText[i])

print(len(TknText))     # Affiche le nombre de phrases de Notre Text

word2count = {}
for data in TknText:
    words = nltk.word_tokenize(data)
    for word in words:
        if word not in word2count.keys():
            word2count[word] = 1
        else:
            word2count[word] += 1

print(word2count)            

import heapq
freq_words = heapq.nlargest(100, word2count, key=word2count.get)

X = []
for data in TknText:
	vector = []
	for word in freq_words:
		if word in nltk.word_tokenize(data):
			vector.append(1)
		else:
			vector.append(0)
	X.append(vector)
X = np.asarray(X)

print(X)
````

````plaintext
import nltk
import numpy as np
import random
import string
import bs4 as bs
import urllib.request
import re

OurText = " Beans. I was trying to explain to somebody as we were flying in, that’s corn. That’s beans. And they were very impressed at my agricultural knowledge. Please give it up for Amaury once again for that outstanding introduction. I have a bunch of good friends here today, including somebody who I served with, who is one of the finest senators in the country, and we’re lucky to have him, your Senator, Dick Durbin is here. I also noticed, by the way, former Governor Edgar here, who I haven’t seen in a long time, and somehow he has not aged and I have. And it’s great to see you, Governor. I want to thank President Killeen and everybody at the U of I System for making it possible for me to be here today. And I am deeply honored at the Paul Douglas Award that is being given to me. He is somebody who set the path for so much outstanding public service here in Illinois. Now, I want to start by addressing the elephant in the room. I know people are still wondering why I didn’t speak at the commencement. "

print(OurText) # Affiche notre Text

TknText = nltk.sent_tokenize(OurText)
for i in range(len(TknText)):
    TknText[i] = TknText[i].lower()
    TknText[i] = re.sub(r'\W', ' ', TknText[i])
    TknText[i] = re.sub(r'\s+', ' ', TknText[i])

print(len(TknText))     # Affiche le nombre de phrases de Notre Text

word2count = {}
for data in TknText:
    words = nltk.word_tokenize(data)
    for word in words:
        if word not in word2count.keys():
            word2count[word] = 1
        else:
            word2count[word] += 1

print(word2count)            

import heapq
freq_words = heapq.nlargest(100, word2count, key=word2count.get)

X = []
for data in TknText:
	vector = []
	for word in freq_words:
		if word in nltk.word_tokenize(data):
			vector.append(1)
		else:
			vector.append(0)
	X.append(vector)
X = np.asarray(X)


print(X) # Affiche Matrice Binaire


word_idf_values = {}
for word in freq_words :
    doc_containing_word=0
    for document in TknText:
        if word in nltk.word_tokenize(document):
          doc_containing_word +=1
    word_idf_values[word] = np.log(len(TknText)/(1+doc_containing_word))

print(word_idf_values) # affich idf valeurs

word_tf_values ={}
for word in freq_words:
  sent_tf_vector = []
  for document in TknText:
    doc_freq=0
    for wrd in nltk.word_tokenize(document):
      if word == wrd:
        doc_freq +=1
    word_tf= doc_freq/len(nltk.word_tokenize(document))
    sent_tf_vector.append(word_tf)
  word_tf_values[word]= sent_tf_vector

print(word_tf_values) #affiche Tf valeurs

tfidf_values= []
for word in word_tf_values.keys():
  tfidf_sentences = []
  for tf_sentences in word_tf_values[word]:
    tf_idf_score = tf_sentences * word_idf_values[word]
    tfidf_sentences.append(tf_idf_score)
    tfidf_values.append(tfidf_sentences)

print(tfidf_values) # affich le resulatat final ..Tf*Idf 


````
